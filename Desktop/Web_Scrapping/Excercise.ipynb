```python
import requests

from bs4 import BeautifulSoup as bs
```

```python
r = requests.get("https://keithgalli.github.io/web-scraping/webpage.html")
soup = bs(r.content) 
# print(soup.prettify())  
```

## Grab all social links from webpage

```python
social_link = soup.find("ul", attrs={"class": "socials"});
social_links =social_link.find_all("a")
print(social_links)
for social_linking in social_links:
    print(social_linking["href"])
```

```python
social_link = soup.select("ul.socials a")
for social_links in social_link:
    print(social_links["href"])
```

## Scrap the Table

```python
stats = soup.find("table", attrs={"class": "hockey-stats"});
# stats
```

```python
import pandas as pd
column = stats.find("thead").find_all("th");
print(column)
column_names = [c.string for c in column]
column_names
```

```python
rows = stats.find("tbody").find_all("tr")
# print(rows[0])
l=[]
for tr in rows:
    table_data = tr.find_all("td")
    row_data = [str(tr.text).strip() for tr in table_data]
    l.append(row_data)
print(l[0]);
```

```python
import pandas as pd
df = pd.DataFrame(l,columns = column_names)
df.head()
```

## Solve the Challenge

```python
files = soup.select('div.block a')
relative_files = [f['href'] for f in files]

url = "https://keithgalli.github.io/web-scraping/"
for f in relative_files:
    full_url = url+f
    page=requests.get(full_url)
    bs_page= bs(page.content)
    secret_word_elem = bs_page.find("p", attrs={"id":"secret-word"})
    secret_word = secret_word_elem.get_text()
    print(secret_word)
```
